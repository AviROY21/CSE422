```python
import random

class Genetic_Algo():
    def __init__(self,p,q,players,player_run,population):
        self.run_prediction=int(q)
        self.player_count=int(p)
        self.players=players
        self.player_run=player_run
        self.population=population
        self.offSprings=[]
        self.newOffSpring=[]
        self.iteration=0

    def offspring_cromosom_create(self):
        if 1<=self.run_prediction or self.run_prediction<=1000 or 1<=self.player_count or self.player_count<=18:

            for j in range(len(self.population)):
                total_run=0
                count=0

                for i in self.population[j]:

                    if i=="0":
                        pass
                    else:
                        a=self.players[count]
                        total_run=total_run+ int(self.player_run[a])
                    count=count+1

                if total_run==self.run_prediction:    #player er run==prediction hole
                    print(self.players)
                    print(self.population[j])
                    return

                elif abs(total_run-int(q))<=10:
                    self.offSprings.append(self.population[j])  #jegula selected oigula dhukabo offspringa
            self.crossover()

    def crossover(self):
        self.newOffSpring=[]
        rand=random.randint(0,self.player_count-1)

        for i in range(len(self.offSprings)-1):
            old1=self.offSprings[i]

            for j in range(i+1,len(self.offSprings),1):
                old2=self.offSprings[j]
                temp=old1[rand]
                temp1=old2[rand]
                new1=old1[0:rand]+temp1+old1[rand+1:self.player_count]   #crossover hoise aikhane
                new2=old2[0:rand]+temp+old2[rand+1:self.player_count]
                self.newOffSpring.append(new1)
                self.newOffSpring.append(new2)

        self.offSprings=self.newOffSpring
        self.mutation()


    def mutation(self):
        a=[]
        rand=random.randint(0,self.player_count-1)

        for i in self.offSprings:
            if i[rand]=="0":
                new=i[0:rand]+"1"+i[rand+1:self.player_count]
            else:
                new=i[0:rand]+"0"+i[rand+1:self.player_count]
            a.append(new)
        self.offSprings=self.offSprings+a
        self.fitnessChecker()

    def fitnessChecker(self):
        a=True
        self.iteration = self.iteration+1

        for j in range(len(self.offSprings)):
            total_run=0
            count=0

            for i in self.offSprings[j]:
                if i=="0":
                    pass
                else:
                    a=self.players[count]
                    total_run=total_run+ int(self.player_run[a])
                    count=count+1

            if total_run==self.run_prediction:
                print(self.players)
                print(self.offSprings[j])
                a=False
                break
        if a:
            if self.iteration==2:
                  print(self.players)
                  print(-1)
            else:
                self.crossover()

txt=open('input.txt','r')
x=txt.read().split("\n")
players=[]
player_run={}    #{"sakib":110}

for i in range(len(x)):
    y=x[i].split(" ")
    if i==0:
        p=y[0]   #8
        q=y[1]   #330
    else:
        players.append(y[0])
        player_run[y[0]]=y[1]   #player num: player run

population=[]

for i in range(pow(2,int(p))):     #2^8
    x=""                           #random value assign korbo
    for j in range(int(p)):
        x= x + str(random.randint(0,1))

    if x not in population:
        population.append(x)

firstPrediction=Genetic_Algo(p,q,players,player_run,population)
firstPrediction.offspring_cromosom_create()
```
